```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
samples_count = 100
total = 10
heads_count = 7
comb_count = math.comb(total, heads_count)

def probability(
    heads_count,
    tosses_count,
    probability_of_head
):
    combs = math.comb(tosses_count, heads_count)
    probH = probability_of_head**heads_count
    probNH = (1-probability_of_head)**(tosses_count - heads_count)
    return combs * probH * probNH

xs = np.array([i / samples_count for i in range(samples_count+1)])
ys = np.array([probability(heads_count, total, x) for x in xs])

def area():
    dx = 1 / samples_count
    return sum([dx * y for y in ys])

# yMultiplier = 1 / area()
# ys *= yMultiplier

print(area())
```

```python
plt.plot(xs, ys)
plt.show()
```

```python
success_rate = 0.8
total_events_count = 50

def combination_probability(
    success_rate, 
    total_events_count, 
    successes_count
):
    number_of_combinations = math.comb(total_events_count, successes_count)
    return number_of_combinations * success_rate**successes_count * (1 - success_rate)**(total_events_count - successes_count)

xs = [successes_count for successes_count in range(total_events_count)]
ys = [combination_probability(success_rate, total_events_count, successes_count)
     for successes_count in xs]

def area():
    return sum(ys)

print(area())
```

```python
plt.plot(xs, ys)
plt.show()
```
