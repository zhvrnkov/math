```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
import math
```

```python

def area(f, a, b, n):
    dx = (b - a) / n
    return sum([f(a + dx * i) * dx for i in n])

def rect(origin, size, ax):
    (x, y) = origin
    (w, h) = size
    r = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(r)
    
def area_rects(f, a, b, n):
    _, ax = plt.subplots()
    dx = (b - a) / n
    for i in range(1, n+1):
        x = a + dx * i
        height = f(x)
        origin = (a + dx * (i - 1), 0)
        size = (dx, height)
        rect(origin, size, ax)
```

```python
def f(x):
    return x**2
vf = np.vectorize(f)
xs = np.linspace(-1, 1, 100)
ys = vf(xs)

area_rects(f, 0, 1, 50)
plt.plot(xs, ys)
```

```python
xs = np.linspace(-math.pi, math.pi, 100)
ys = np.sin(xs)
area_rects(math.sin, 0, math.pi, 50)
plt.plot(xs, ys)
```

```python
def f2(x):
    return math.exp(-(x**2))
a = -3
b = 3
vf2 = np.vectorize(f2)
xs = np.linspace(a, b, 100)
ys = vf2(xs)
area_rects(f2, a, b, 100)
plt.plot(xs, ys)
```
