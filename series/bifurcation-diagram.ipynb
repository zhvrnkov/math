```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
%matplotlib notebook
```

```python
f = lambda x, r: r * x * (1 - x)

def get_ys(r, x0, count):
    x = x0
    ys = [x]
    for _ in range(count-1):
        val = f(x, r)
        x = val
        ys.append(val)
    return ys

def plot(x0, r, count):
    xs = range(count)
    ys = get_ys(r, x0, count)
    plt.plot(xs, ys)
    plt.text(xs[-1], ys[-1], "%.1f" % r, fontsize=8)
    
def dl(upper):
    return np.linspace(0, upper, upper * 10 + 1)
   
fig = plt.figure()
for r in dl(2):
    for x0 in dl(1):
        plot(x0, r, 100)
        
```

```python
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

count = 100
xs = range(count)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

line, = ax.plot(xs, get_ys(0, 0, count))

def update(r, x0):
    line.set_ydata(get_ys(r, x0, count))
    fig.canvas.draw_idle()

interact(update, r=(0, 3, 0.1), x0=(0, 1, 0.01))
```

```python
x0 = 0.51

count = 500

def get_rs(m):
    return np.linspace(0, m, int(m * 1000 + 1))

def get_ps(rs):
    return [get_ys(r, x0, count)[-1] for r in rs]

def get_ps2(rs):
    return [get_ys(r, x0, count)[-2] for r in rs]


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.margins(2, 2)
rs = get_rs(2)
line, = ax.plot(rs, get_ps(rs))
line2, = ax.plot(rs, get_ps2(rs))

def update(m):
    rs = get_rs(m)
    line.set_xdata(rs)
    line.set_ydata(get_ps(rs))
    line2.set_xdata(rs)
    line2.set_ydata(get_ps2(rs))
    fig.canvas.draw_idle()
    
interact(update, m=(0, 4, 0.001))
```
