```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
L = 1.0
dt = 0.01
thetas = [1.0]
theta_vs = [0.0]

def g():
    return 9.8

def theta_a(theta, theta_v):
    air_resistance = 0.0
    return -air_resistance * theta_v - g() / L * sin(theta)

for i in range(1000):    
    a = theta_a(thetas[-1], theta_vs[-1])
    theta_vs.append(theta_vs[-1] + a * dt)
    thetas.append(thetas[-1] + theta_vs[-1] * dt)

plt.plot(thetas, theta_vs)
```

```python
space = np.mgrid[-10:10:0.5, -10:10:0.5].reshape(2,-1).T
xs = []
ys = []
dxs = []
dys = []

scale_down = 0.01

def vec(theta, theta_v):
    a = theta_a(theta, theta_v)
    return [theta_v + a, ]

for [x, y] in space:
    a = theta_a(x, y)
    v = y + a

    xs.append(x)
    ys.append(y)
    dxs.append(v * scale_down)
    dys.append(a * scale_down)

plt.quiver(xs, ys, dxs, dys)
```
