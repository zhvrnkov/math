```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
%matplotlib notebook
```

```python
def draw(input, func):
    y_origin = 0
    y_distance = 1
    xs = [[x, func(x)] for x in input]
    ys = [[y_origin + y_distance, y_origin] for x in input]
    plt.plot(xs, ys, marker = 'o')
    lines = []
    for i in range(len(xs)):
        lines.append(xs[i])
        lines.append(ys[i])
    plt.plot(*lines)
    plt.xlabel("X axis label")
    plt.ylabel("Y axis label")
    plt.show()
    
inputs = np.linspace(1, 100, 10)
func = lambda x: x ** 2

draw(inputs, func)
```

In this representation of a function's graph we plot xs and the top, plt ys at the bottom. 

So any function is basically transforms number line.

From such point of view, the derivatives are also interesting. Consider some x and its two neighbors: x1 = x - dx/2, x2 = x + dx/2. x is mapped to f(x) = y, x1 to y1, x2 to y2. We can consider this x2 <-> x1 as a line region, which will spread out or contract in transformed region y2 <-> y1. df/dx = (y2 - y1) / (x2 - x1).

Also some facts of derivative that makes good presentation in this approach:
- df/dx < 1 => (x2 - x1) > (y2 - y1) => line is contracted
- df/dx > 1 => line is spread out
- df/dx < 0 => (x2, y2) and (x1, y1) lines intersects
